```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Results of NSAID2.csv')
```

```python
# Count of gender
#sns.countplot(x='GENDER', data=df)
#plt.title('Gender Distribution')
#plt.show()

# Age distribution
plt.hist(df['AGE(years)'], bins=10, color='blue', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()




```

```python
# Count of gender
SELF_MEDICATION_OR_NOT_counts = df['SELF MEDICATION OR NOT?'].value_counts()
print(SELF_MEDICATION_OR_NOT_counts)

sns.countplot(y='SELF MEDICATION OR NOT?', data=df)
plt.title('Self-Medication Distribution')

# Print the count of follow-ups
print("Self medication Counts:")


plt.show()
```

```python
# Occupation distribution
sns.countplot(y='OCCUPATION', data=df)
plt.title('Occupation Distribution')
plt.show()
OCCUPATION_counts = df['OCCUPATION'].value_counts()
print(OCCUPATION_counts)
```

```python
# Combine the active ingredients of drug 1 and drug 2 into a single 
# Combine the active ingredients of drug 1 and drug 2 into a single column
df['ACTIVE INGREDIENTS'] = df['ACTIVE INGREDIENT OF DRUG 1 '] + ', ' + df['ACTIVE INGREDIENT OF DRUG 2']

# Count the frequency of each active ingredient
drug_counts = df['ACTIVE INGREDIENTS'].value_counts()

# Select the top N most consumed drugs (adjust N as needed)
top_n = 5
top_drugs = drug_counts.head(top_n)

# Create a bar plot to show the most consumed drugs
plt.figure(figsize=(10, 6))
top_drugs.plot(kind='bar', color='skyblue')
plt.xlabel('Active Ingredients')
plt.ylabel('Frequency')
plt.title(f'Top {top_n} Most Consumed Drugs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count the frequency of each dosage form
dosage_counts = df['DOSAGE FORM'].value_counts()

# Create a bar plot to show the most consumed dosage forms
plt.figure(figsize=(10, 6))
dosage_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Dosage Form')
plt.ylabel('Frequency')
plt.title('Most Consumed Dosage Forms')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print(dosage_counts)


```

```python
# Count the frequency of each side effect
side_effects_counts = df['SIDE EFFECTS'].value_counts()

# Select the top N side effects (adjust N as needed)
top_n = 10
top_side_effects = side_effects_counts.head(top_n)

# Create a horizontal bar chart to show the distribution of top side effects
plt.figure(figsize=(10, 8))
top_side_effects.sort_values().plot(kind='barh', color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Side Effects')
plt.title(f'Top {top_n} Side Effects')
plt.tight_layout()
plt.show()


print(side_effects_counts)
```

```python
 # Create a crosstab of occupation and source of information
occupation_source_crosstab = pd.crosstab(df['OCCUPATION'], df['SOURCE OF INFORMATION'])

# Create a grouped bar chart
plt.figure(figsize=(10, 8))
occupation_source_crosstab.plot(kind='bar', stacked=True, cmap='Paired')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Relationship between Source of Information and Occupation')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title='Source of Information')
plt.show()



```

```python
# Calculate statistics for males
males = df[df['GENDER'] == 'Male']
total_males = len(males)
#average_children_males = males['children'].mean()

# Calculate statistics for females
female_data = df[df['GENDER'] == 'Female']
total_females = len(female_data)
#average_children_females = female_data['children'].mean()

# Calculate statistics for children
#total_children = df['children'].sum()
#average_children_all = df['children'].mean()
```

```python
labels = ['Males', 'Females']
sizes = [total_males, total_females]
colors = ['blue', 'pink']
explode = (0.1, 0,)  # explode 1st slice (Males)

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Distribution of Males & Females")

# Display the pie chart
plt.show()

print(total_males)
print(total_females)
```

```python
# Count the occurrences of each value in the 'Follow-up' column
follow_up_counts = df['Follow-up'].value_counts()

# Plot the frequency of follow-ups using a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=follow_up_counts.index, y=follow_up_counts.values)
#plt.xticks(rotation=0)
plt.xlabel('Follow-up')
plt.ylabel('Count')
plt.title('Frequency of Follow-ups')
plt.tight_layout()

# Display the bar plot
plt.show()

# Print the count of follow-ups
print("Follow-up Counts:")
print(follow_up_counts)
```
